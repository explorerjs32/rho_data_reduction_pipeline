```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import os
from astropy.visualization import ZScaleInterval, ImageNormalize
from astropy.stats import SigmaClip
from photutils.background import Background2D, MedianBackground
from photutils.segmentation import detect_threshold, detect_sources
from photutils.utils import circular_footprint
```

```python
# Define the data directory
data_dir = '../data/2024-04-15/'

# Create a list of images that correspond to light frames
image_data_list = [fits.getdata(os.path.join(data_dir, file)) 
                  for file in os.listdir(data_dir)
                  if file.startswith('RRGeminorum')
                  if fits.getheader(os.path.join(data_dir, file))['FRAME'] == 'Light']
```

```python
# Define the paameters to estimate the the sky background of the image
# A non-uniform background brightness is going to be assumed
sigma_clip = SigmaClip(sigma=3.0, maxiters=10)
threshold = detect_threshold(image_data_list[0], nsigma=2.0, sigma_clip=sigma_clip)
segment_img = detect_sources(image_data_list[0], threshold, npixels=10)
footprint = circular_footprint(radius=10)
mask = segment_img.make_source_mask(footprint=footprint)

box_size = (30, 30)
filter_size = (3, 3)
bkg_estimator = MedianBackground()


# Estimate the 2D background of the image
bkg = Background2D(image_data_list[0], box_size=box_size, mask=mask, filter_size=filter_size, 
                   sigma_clip=sigma_clip, bkg_estimator=bkg_estimator)

fig, ax = plt.subplots(2,2, figsize=(12,8), tight_layout=True)
ax[0,0].imshow(image_data_list[0], origin='lower', cmap='gray', norm=ImageNormalize(image_data_list[0], interval=ZScaleInterval()))
ax[0,1].imshow(image_data_list[0], origin='lower', cmap='gray', norm=ImageNormalize(image_data_list[0], interval=ZScaleInterval()))
bkg.plot_meshes(ax=ax[0,1], marker='.',outlines=True, color='cyan', alpha=0.3)
ax[1,0].imshow(bkg.background, origin='lower', cmap='gray')
ax[1,1].imshow(image_data_list[0] - bkg.background, origin='lower', cmap='gray', 
             norm=ImageNormalize(image_data_list[0] - bkg.background, interval=ZScaleInterval()))

ax[0,0].set_title('Raw Image')
ax[0,1].set_title('Segment to Estimate Background')
ax[1,0].set_title('Estimated Background')
ax[1,1].set_title('Background Subtracted Image')


plt.show()
```

```python
print(f"Background Median: {bkg.background_median} [ADU/pixel]\nBackground RMS: {bkg.background_rms_median} [ADU/pixel]")
```

```python
# Define the function to implement into the main code
def background_subtraction(image):

    # Define the paameters to estimate the the sky background of the image
    # A non-uniform background brightness is going to be assumed
    sigma_clip = SigmaClip(sigma=3.0, maxiters=10)
    threshold = detect_threshold(image, nsigma=2.0, sigma_clip=sigma_clip)
    segment_img = detect_sources(image, threshold, npixels=10)
    footprint = circular_footprint(radius=10)
    mask = segment_img.make_source_mask(footprint=footprint)
    box_size = (30, 30)
    filter_size = (3, 3)
    bkg_estimator = MedianBackground()
    
    # Estimate the 2D background of the image
    bkg = Background2D(image, box_size=box_size, mask=mask, filter_size=filter_size, 
                       sigma_clip=sigma_clip, bkg_estimator=bkg_estimator)

    # Subtract the background from the image reduced image
    bkg_subtracted_image = image - bkg.background

    fig, ax = plt.subplots(2,2, figsize=(12,8), tight_layout=True)
    ax[0,0].imshow(image, origin='lower', cmap='gray', norm=ImageNormalize(image, interval=ZScaleInterval()))
    ax[0,1].imshow(image, origin='lower', cmap='gray', norm=ImageNormalize(image, interval=ZScaleInterval()))
    bkg.plot_meshes(ax=ax[0,1], marker='.',outlines=True, color='cyan', alpha=0.3)
    ax[1,0].imshow(bkg.background, origin='lower', cmap='gray')
    ax[1,1].imshow(image - bkg.background, origin='lower', cmap='gray', 
                 norm=ImageNormalize(image - bkg.background, interval=ZScaleInterval()))
    
    ax[0,0].set_title('Raw Image')
    ax[0,1].set_title('Segment to Estimate Background')
    ax[1,0].set_title('Estimated Background')
    ax[1,1].set_title('Background Subtracted Image')
    
    plt.show()
        
    return bkg_subtracted_image

for image in image_data_list:
    background_subtraction(image)
```
